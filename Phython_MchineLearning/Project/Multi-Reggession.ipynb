```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('Startups.csv')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
dat = pd.factorize(df['State'])
df['State'] = pd.Series(dat[0])
```

```python
df
```

```python
x1 = df.iloc[:,:-4].values
x2 = df.iloc[:,:-3].values
x3 = df.iloc[:,:-2].values
x4 = df.iloc[:,:-1].values
y = df.iloc[:,4].values
```

```python
plt.scatter(x1,y)
```

```python
y.shape
```

```python
model1 = LinearRegression()
model1.fit(x3,y)
```

```python
print(model1.intercept_)
```

```python
print(model1.coef_)
```

```python
x3 = sm.add_constant(x3)
```

```python
model = sm.OLS(y,x3).fit()
model.summary()
```
